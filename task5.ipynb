```python
df2 = pd.DataFrame({'key':['b','b','a','c','a','a','b'],
                   'data1':range(7)})
df3 = pd.DataFrame({'key':['a','b','d'],
                   'data2':range(3)})
display(df2,df3)
```

```python
pd.merge(df2,df3)
```

```python
df4 = pd.DataFrame({'key1':['b','b','a','c','a','a','b'],
                  'data1':range(7)})
df5 = pd.DataFrame({'key2':['a','b','d'],
                   'data2':range(3)})
display(df4,df5)
```

```python
pd.merge(df4,df5,left_on='key1',right_on='key2')
```

```python
pd.merge(df4,df5)
```

```python
pd.merge(df2,df3,how='outer')
```
